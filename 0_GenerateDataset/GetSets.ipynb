```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc

import os
```

```python
#import exp_filtered
data = pd.read_csv('../Datasets/experiments_valid.csv',header=0, index_col=0)
experiments = pd.DataFrame(data)

experiments.tail(10)
```

```python
sets=experiments.drop(columns=['experiment','start','end','startUnix','endUnix','topology']).reset_index(drop=True).drop_duplicates()

sets.head(10)
```

```python
sets['initialDelay'] = 100
sets = sets[['d','dlo','dhi','dscore','dlazy','dout','gossipFactor','initialDelay','interval','parameter']].sort_values(by=['parameter'])

sets.head(100)
```

```python
# final = pd.concat[sets.reset_index(drop=True), sets.reset_index(drop=True), sets.reset_index(drop=True)]
final = pd.DataFrame(np.repeat(sets.values, 3, axis=0), columns=sets.columns)

final = final.sort_values(by=['parameter'])

final.head(10)
```

```python
final.to_csv('../Datasets/parameters.csv', header=False, index=False)
```
