```python
#Aggregate datasets for netwise view

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc

import os
```

```python
basepath='/home/flav/gossip/causalGossipSub/Datasets/1node/'

consolidated = pd.DataFrame()
directory = os.fsencode(basepath)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    path = os.path.join(basepath, filename)
    if os.path.isdir(path):
        pass
    else:
        data = pd.read_csv(basepath+filename,header=0, index_col=0)
        aux = pd.DataFrame(data)
    
        consolidated = pd.concat([consolidated, aux])
    
consolidated.head(25)
     
```

```python
consolidated.to_csv('../Datasets/Networkwise/consolidated_5s.csv')
```
