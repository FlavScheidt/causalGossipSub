```python
unls = pd.DataFrame()
directory = os.fsencode('/home/flav/gossip/gossipGoSnt/clusterConfig/unl/')

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    unl =  Path(filename).resolve().stem
    with open("/home/flav/gossip/gossipGoSnt/clusterConfig/unl/"+filename, 'r') as fp:
        x = len(fp.readlines())
    aux = pd.DataFrame({'unl': [unl],'size': [x]})
    unls = pd.concat([unls, aux])
    

unls.head(10)
```

```python
unlSize = df.merge(unls, on='unl').drop(columns=['unl'])
unlSize = unlSize.rename(columns={'size':'unl'})

unlSize.head(24)
```

```python
#Validator
validatorsSize=unlSize
directory = os.fsencode('/home/flav/gossip/gossipGoSnt/clusterConfig/validator')

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    validators =  Path(filename).resolve().stem
    with open("/home/flav/gossip/gossipGoSnt/clusterConfig/validator/"+filename, 'r') as fp:
        x = len(fp.readlines())
    validatorsSize.loc[validatorsSize['node'] == validators, 'validator'] = str(x)
    
validatorsSize.head(25)
     
```

```python
topology = validatorsSize

topology['general'] = 24
topology.to_csv('../Datasets/topologySize.csv', index=False)
```
